```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_excel('0.xls')

plt.scatter(data.X, data.Y)
plt.plot([-1, 3], [4, 8], c = '#17becf')
plt.show()

above_line = (data.Y - 5) - data.X > 0
cluster1 = data[above_line]
cluster2 = data[np.logical_not(above_line)]
plt.scatter(cluster1.X, cluster1.Y)
plt.scatter(cluster2.X, cluster2.Y)
plt.plot([-1, 3], [4, 8], c = '#17becf')
plt.show()

cluster1 = cluster1.values
min_sum = np.inf
for i in range(cluster1.shape[0]):
  cur_sum = 0
  for j in range(cluster1.shape[0]):
    cur_sum += np.linalg.norm(cluster1[i] - cluster1[j])
  if cur_sum < min_sum:
    centre1 = cluster1[i]
    min_sum = cur_sum

cluster2 = cluster2.values
min_sum = np.inf
for i in range(cluster2.shape[0]):
  cur_sum = 0
  for j in range(cluster2.shape[0]):
    cur_sum += np.linalg.norm(cluster2[i] - cluster2[j])
  if cur_sum < min_sum:
    centre2 = cluster2[i]
    min_sum = cur_sum

print((centre1 + centre2) * 1000 / 2)
```
